```python
brand=[]
model=[]
manu_year=[]
driven=[]
fuel=[]
owners=[]
location=[]
price=[]
gear=[]
```

```python
#for maharashtra
for i in page_url:
    
    driver.get(i)
    try:
        a=driver.find_element_by_xpath("//div[@class='_35xN1']")

        brand.append(a.text)
    except:
        brand.append('-')
        
    try:
        b=driver.find_element_by_xpath("//div[@class='_3tLee']")

        model.append(b.text)
    except:
        model.append('-')
    
    try:
        c=driver.find_element_by_xpath("//div[@class='_1l939']/div[1]/div[2]/div[2]")

        owners.append(c.text)
    except:
        
        owners.append('-')
        
        
    try:
        e=driver.find_element_by_xpath("//div[@class='aOxkz']/div[1]/div[1]")

        fuel.append(e.text)
    except:
        fuel.append('-')
        
    try:
        f=driver.find_element_by_xpath("//div[@class='aOxkz']/div[2]/div[1]")

        driven.append(f.text)
    except:
        driven.append('-')   
    
    try:
        g=driver.find_element_by_xpath("//div[@class='aOxkz']/div[3]/div[1]")

        gear.append(g.text)
    except:
        gear.append('-') 
        
    try:
        h=driver.find_element_by_xpath("//div[@class='_3FkyT']")

        price.append(h.text)
    except:
        price.append('-') 
```

```python
brand_=[]
for i in range(len(brand)):
    u=brand[i].split('(')[0]
    brand_.append(u)
brand_
```

```python
for i in range(len(brand)):
    l=brand[i].split('(')[-1].split(')')[0]
    manu_year.append(l)
manu_year
```

```python
price_=[]
for i in range(len(price)):
    po=price[i].split('₹')[-1]
    price_.append(po)
price_
```

```python
df1 = pd.DataFrame({})
df1["Brand"] = brand_
df1["Model"] = model
df1["Mnf_year"] = manu_year
df1["Driven_kms"] = driven
df1["Fuel"] = fuel
df1["Gear"] = gear
df1["Owners"]=owners
df1["Car_Price"] = price_
df1.insert(0, "Location", ["Maharashtra"]*500, True)
df1
```
